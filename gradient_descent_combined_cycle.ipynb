```python
def cost(points,M):
    total_cost=0
    N=points.shape[0]
    for i in range(N):
        x=points[i,:-1]
        x=np.append(x,1)
        y=points[i,-1]
        total_cost+=(1/N)*(y - (M * x).sum()) ** 2
    return total_cost
```

```python
def step_gradient_descent(points,learning_rate,M):
    cols=points.shape[1]
    new_M=np.zeros(cols)
    N=points.shape[0]
    for i in range(N):
        x=points[i,:-1]
        x=np.append(x,1)
        y=points[i,-1]
        for j in range(cols):
            new_M[j] += (-2/N) * (y - (M*x).sum())*x[j]
        M=M-(learning_rate*new_M)
    return M
    
```

```python
def gd(points,num_iterations,learning_rate):
    cols=train_data.shape[1]
    M=np.zeros(cols)
    for i in range(num_iterations):
        M=step_gradient_descent(points,learning_rate,M)
        print(i,"cost : ",cost(points,M))
    return M
```

```python
def run():
    num_iterations=100
    learning_rate=0.0000001
    m=gd(train_data,num_iterations,learning_rate)
    return m[:-1],m[-1]
```

```python
m,c=run()
print(m,c)
```

```python
def predict(x,m,c):
    N=x.shape[0]
    y=np.zeros(N)   # gives an array with all zeroes of size 127 since shape of x is (127,13) 
    for i in range(N):
        x_points=x[i,:]
        y[i]=(m*x_points).sum()+c
    return y
    
y_predict=predict(test_data,m,c)
print(y_predict.shape)
y_predict
```
